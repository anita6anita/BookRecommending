```python
#merging ratings and books based on ISBN column
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
#counting the no.of ratings for each book
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
#changing datatype of book-rating col to int to perform aggregate functions
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')
```

```python
#books with avg rating
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'}, inplace=True)
avg_rating_df
```

```python
#merging no.of ratings and average ratings
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
#taking top 250 books with highest average rating
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
popular_df
```

```python
#displaying popular df with specified columns
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
popular_df
```

```python
#displaying users index who have read more than 250 books
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
reading_users = x[x].index
reading_users
```

```python
#displaying the reading users with specified columns from merged table using ratings and books df
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(reading_users)]
filtered_rating
```

```python
#taking books with no.of ratings more than 50
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
famous_books
```

```python
#displaying famous books with columns with filter ratings
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
#creating pivot table with index,columns and values as specified
pt = final_ratings.pivot_table(index='User-ID',columns='Book-Title',values='Book-Rating')
pt
```

```python
#filling empty values with 0's
pt.fillna(0,inplace=True)
```
